### Differences between ndarray and lists

* lists can contain *multiple types* in one list. ndarrays are of *one type only*.
 * No type checking necessary -> more efficient
* Accessing: 
 * list[x][y] vs. array[x,y]
 * list[:1][:1] does not work, vs. array[:1,:1] does work.
* universal functions on arrays work efficiently: np.sin( array )

### Example: student grades

Consider the following 2 arrays:
  * 1-d array with student names
  * 2-d array with multiple grades on rows. Each row contains the grades
  of one student.

````python
# Generate some names and grades.
student = np.array(['Anne', 'Bob', 'Mary', 'John'])
grades = 6 * np.random.random(size=(4, 8)) + 4
grades
student
````

````python
# Show the grades of Bob
bob_grades =  grades[ student == 'Bob' ]
bob_grades
````

````python
# Note, the answer is still a 2d array (1,8) 
bob_grades.shape
````

````python
# Enumerate over the entries in the array
for (nr, g) in enumerate( bob_grades[0] ):
    print( "Grade %d is %.1f" % (nr+1, g) )
````

````python
# Show the grades of Anne and Mary
# Use parenthesis! 
# or = |
# and = &
grades [ (student == 'Anne') | (student == 'Mary') ]
````

````python
# Show all the grades that did not pass
grades[grades < 5.8]
````

````python
# Show the names who have received a 9.0 or higher at least once

# First, which values are >= 9
grades >= 9.0
````

````python
# Now for each row, which ones have a True at least once...
np.any( grades > 9, axis=1)
````

````python
# Now use this as a boolean index into the student names.
student[ np.any( grades > 9, axis=1) ]
````

````python
# Show the min, max, and average grade
print("Min", grades.min())
print("Max", grades.max())
print("Mean", grades.mean())
````

````python
# Bob's average
grades[ student == 'Bob' ].mean()
````

````python
# Averages per student
# mean() can take an argument to take the mean by collapsing a given axis (0=row, 1=column)
# This means: axis=0 - calculate mean per column, axis=1 - calculate men per row.
# No argument: average of all values.
# max and min can take the same argument.
grades.mean( axis=1 )
````

````python
for (name, grade) in zip( student, grades.mean( axis=1 )):
    print('Average of %s is %.1f' % (name, grade) )
````

````python
# Average grade per test (mean per column),
# so we can see which test was made the most difficult.
grades.mean( axis = 0 )
#grades.mean(axis = 0) == grades.mean(axis = 0).min()
````

````python
# Quick peek: Let's plot the averages
import matplotlib.pyplot as plt
plt.bar( np.arange( grades.shape[1] ), grades.mean( axis = 0 ) )
````

````python
# Or plot the averages of each student
plt.bar( np.arange( grades.shape[0]), grades.mean( axis = 1), align='center', color='red' )

# Nice labels on the x-axis
plt.xticks( np.arange( len(student) ), student)

# Label the axis
plt.xlabel("Student")
plt.ylabel("Grade")

# Put the actual grade on top of the bars, for clarity
for (i, grade) in enumerate( grades.mean(axis=1)):
    plt.text( i - 0.13, grade + 0.25, "%.2f" % (grade) )
````

### Example: Random Walks

Consider a random walk, represented by a sequence of steps (1 or -1) occuring with equal
probability. Let's look at the walks when starting at 0 and doing 1000 steps.

  * How far from 0 did we get (positive, negative)?
  * When doing 5000 walks, how many of those reached at least 30/-30?
  * How many steps did that take on average?
  


````python
# A walk is represented by an array with 1 and -1.
nsteps = 1000

# Generate an array of 0 and 1.
steps = np.random.randint( 2, size=nsteps)
steps[:10]
````

````python
# Now replace all 0's with -1, leave the others alone.
steps = np.where(steps==0, -1, steps)
````

````python
# Let's peak what is inside steps
steps[:10]
````

````python
# Now, turn these steps into a real walk
walk = steps.cumsum()

walk[:10]
````

````python
# Find the minimum and maximum positions of this walk
print("Minimum of walk =", walk.min())
print("Maximum of walk =", walk.max())
````

````python
# A quick plot
plt.plot( np.arange(nsteps), walk)
````

````python
# Where did this walk cross the 30/-30 line?

# Only look at the absolute value of the walk. Then compare it to 30.
walk30 = np.abs(walk) >= 30

walk30
````

````python
# Now, find the first occurrence of True
walk30.argmax()
````

````python
# We can store 5000 walks of 1000 steps each in a 2d matrix.
# One walk per row.

# First create a matrix with 5000 rows of 1000 steps
nsteps=1000
nwalks=5000

steps = np.random.randint( 2, size=(nwalks, nsteps))
steps = np.where(steps == 0, -1, steps)
````

````python
# Show a small subset
steps[:3,:10]
````

````python
# Turn the steps into walks
walks = steps.cumsum(axis=1)
walks[:3,:10]
````

````python
# What is the maximum/minimum over all walks?
print("Minimum", walks.min())
print("Maximum", walks.max())
````

````python
# Minimum per walk?
walks.min(axis=1)
````

````python
# How many walks reached at least 30/-30?

# First, only look at absolute values: np.abs().
# Then, check those values with >= 30

reached30 = np.abs(walks) >= 30

reached30
````

````python
# How many of those rows contain at least one True value?

# This will yield an array with 5000 booleans. True if row contains at least one True.
walks30 = np.any( reached30, axis=1)

walks30
````

````python
# Now, count the number of True values... that is the number of walks that have
# reached 30/-30

# When summing booleans, True = 1 and False = 0.
walks30.sum()
````

````python
# Now find the index when a walk crossed the 30/-30 boundary first?
# The first-crossing.

reached30.argmax(axis=1)

# First, remove all walks that did not reach 30 (so the mean value is corect later)
selected_walks = reached30[ walks30 ]

# Now, find the first "True" value in these selected_walks. That is the first-crossing.
crossing_index = selected_walks.argmax( axis=1 )

crossing_index
````

````python
# What is the average where the walks have crossed 30/-30?

crossing_index.mean()
````

### Sorting and unique

There are 2 ways to sort an array:
  * array.sort() - sorts the array in-place
  * np.sort( array ) - returns a new sorted array
  
  * np.unique( array ) - returns a new sorted array with unique elements only.
  

````python
a = np.random.randint( 10, size=(3,4) )
a
````

````python
# With multi-dimensional arrays, think about the axis along which to sort.
# axis = None : flatten the array
# if axis not specified, sorted along the last axis.
np.sort( a, axis=None )
````

````python
# Sort along row-axis, so sort the values in each "column"
np.sort( a, axis=0)
````

````python
# This is how to sort per row.
np.sort( a, axis=1 )
````

````python
# Find all unique elements in array.
np.unique( a )
````
