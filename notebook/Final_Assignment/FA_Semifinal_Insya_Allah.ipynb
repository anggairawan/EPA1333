# Preprocessing

## World Bank Data Interpolation

Many missing data are found in World Bank raw data. Thus, interpolation is performed to fill those missing data with appropriate values. We need to drop header and indicator of the dataframe so that we create a temporary dataframe with only numerical value of the World Bank data. We need this transformation to accomplish the interpolation.

```python
# create column of years of the world bank data
column_start_of_year = 4
interval_duration = 2016-1960 + 1
years_column = wb.columns.values[range(column_start_of_year,column_start_of_year+interval_duration)]
years_column

# copy the content of the dataframe (to avoid copying the reference) for the interpolation
wb_prep = pd.DataFrame()
wb_prep = wb[years_column]

# reset the column (year) to integer
wb_prep.columns = range(0,len(years_column))

# display the dataframe
wb_prep.head(8)
```

As we can see from dataframe above, we have a particular row with NaN as its value, both partially or fully for entire row. We will perform the interpolation only to the partial one, see row 7 for an example.

```python
# interpolate the data in row direction also in both direction within range limit
wb_interpolated = wb_prep.interpolate(method='linear',limit_direction='both',limit=len(years_column)-1,axis=1)
```

```python
# show the interpolation results
wb_interpolated.head(8)
```

The row now are filled out by values from the interpolation results. As for the case where the entire row is completely fill out by NaN, we will handle it later. 

A new dataframe is created to accomodate the new interpolated values. 

```python
# create a new dataframe to avoid overwriting the original dataframe
wb_inter = pd.DataFrame()
wb_inter = wb[wb.columns.values]

# replace wb_inter annual value with the interpolated one
# now we have interpolated world bank data
wb_inter[list(wb.columns.values[range(column_start_of_year,column_start_of_year+interval_duration)])] = wb_interpolated[list(range(0,len(years_column)))]
```

## Data Cleaning

### Removing Missing Indicator (NaN for entire row)

Check the data for each country which the entire years of a indicator are missing. We will remove this indicator from the dataframe so that it will not corrupt the model.

```python
import operator

# initialised empty list of to be deleted indicator/ missing indicator/  NaN value
list_deleted_indi = []

# initialised empty list of tuple country: number of missing indicator
list_tuple_country_del_indi = []

# number of the indicator as mentioned on metadata
number_of_indicators = len(wb_meta_indi)

# Jump for every country section (index skip 80 rows for each iteration)
for i in range(0,len(df_ccpi),number_of_indicators):
    
    # append the tuple of country name on the section and the number of NaN rows on the country section
    list_tuple_country_del_indi.append((df_ccpi['Country Name'][i], df_ccpi[i:i+number_of_indicators].isnull().any(axis=1).sum())) 
    
    # filter the indicator with NaN rows
    for c in df_ccpi[i:i+number_of_indicators]['Indicator Name'][df_ccpi[i:i+number_of_indicators].isnull().any(axis=1)]:
        
        # append it to the list
        list_deleted_indi.append(c)    

# create a dataframe of country with its missing indicator number
df_tuple_cdi = pd.DataFrame(list_tuple_country_del_indi, columns = ['Country Name','Number of Missing Indicator'])
# set index to country name for plot purpose
df_tuple_cdi = df_tuple_cdi.set_index('Country Name')
```

## Feature Engineering

### Averaging and Summing the Features

As our second question only needs to know which factors that has most significant contribution to climate change, the World Bank needs to be rearranged so the indicators does not represent only for each countries but for whole worlds. Thus, we need to 'engineer' the data. The data will be summed up or re-averaged based on the source note on the countries metadata. The data might be even removed if the data is considered as redundant or are already represented by another feature. The information of the metadata is concluded as the csv file below.

```python
# read feature interaction category 
df_feature_select = pd.read_csv('world_bank/feature_engineering.csv',sep=',', header=0) 
```

```python
# display the 
df_feature_select.head()
```

For example, Urban population (% of ) is removed because this indicator is already represented by Urban Polulation. Meanwhile, Urban Population is summed to obtain the total global urban population. And Urban population growth (annual %) is averaged because the value is stated as percentage and also to obtain global average urban population growth.

```python
# merge the category on the ccpi dataframe
df_ccpi_feature_label = pd.merge(df_ccpi_clean_reset, df_feature_select, 
                                 on = 'Indicator Name', how="inner")

print('Dataframe shape:', df_ccpi_feature_label.shape)
```

Code snippet below will remove indicators labelled with "removed" for CCPI countries data features.

```python
# remove the feature with remove category
df_ccpi_feature_label_removed = df_ccpi_feature_label [df_ccpi_feature_label['Feature Interaction'] != 'remove']
print('Dataframe shape:', df_ccpi_feature_label_removed.shape)
```

The number of features to be summed is 14.

```python
# count the feature with sum as the category 
set_indi_sum = set(df_ccpi_feature_label_removed[df_ccpi_feature_label_removed['Feature Interaction'] == 'sum']['Indicator Name'])
len(set_indi_sum)
```

The number of features to be averaged is 29.

```python
# count the feature with average as the category 
set_indi_average = set(df_ccpi_feature_label_removed[df_ccpi_feature_label_removed['Feature Interaction'] == 'average']['Indicator Name'])
len(set_indi_average)
```

New engineered dataframe is generated. First the dataframe with the averaged features is appended. 

```python
# create a new dataframe with the averaged features as the values
df_ccpi_feature_processed = pd.DataFrame()

# iterate for each to be averaged indicator
for indi in set_indi_average:
    
    # filter the df which the indicator is indi
    df_indi = df_ccpi_feature_label_removed[(df_ccpi_feature_label_removed['Indicator Name'] == indi)]
    
    # sum the feature from each country
    df_indi_ave = (df_indi.sum(axis = 0)) 
    
    # average the values of each feature
    df_ccpi_feature_processed[indi] = df_indi_ave[4:-1]/len(ccpi_country)
    
print('Dataframe shape:',df_ccpi_feature_processed.shape)
```

The summed features are calculated and appended.  

```python
# concat the new dataframe with the summed features
for indi in set_indi_sum:
    
    # filter the df which the indicator is indi
    df_indi = df_ccpi_feature_label_removed[(df_ccpi_feature_label_removed['Indicator Name'] == indi)]
    
    # sum the feature from each country
    df_indi_sum = df_indi.sum(axis = 0)
    
    # copy the values of summed features
    df_ccpi_feature_processed[indi] = df_indi_sum[4:-1]
    
print('Dataframe shape:',df_ccpi_feature_processed.shape)    
```

# Training Set

In this step, input feature matrix and target vector are established from previously loaded data to train the model.

```python
# copy the processed ccpi dataframe to input features 
X_tempe_train = df_ccpi_feature_processed[df_ccpi_feature_processed.columns.values]

# copy the input feature to the temperature dataset
dataset_tempe = X_tempe_train[X_tempe_train.columns.values]

print('Dataframe shape:', dataset_tempe.shape)
# dataset_tempe.head()
```

## Generating the Model

Three different regression model are generated to predict the anomaly of global temperature.

#### Linear Regression

```python
# import the linear regression library
from sklearn.linear_model import LinearRegression

# set intercept to True
lr_tempe = LinearRegression(fit_intercept=True)

# build the model
lr_tempe.fit(X_tempe_train, y_tempe_train)
```

#### Random Forest Regression

```python
# import Random Forest Regresson library
from sklearn.ensemble import RandomForestRegressor

# set the max depth of the tree
max_depth = 30

# initialised the model
regr_rf = RandomForestRegressor(max_depth=max_depth, random_state=2)

# build the model
regr_rf.fit(X_tempe_train, y_tempe_train)
```

#### Support Vector Regression

```python
# import Support Vectro Regressor library
from sklearn.svm import SVR

# set the kernel, hardness, and gamma
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)

# build the model
svr_rbf.fit(X_tempe_train, y_tempe_train)
```

## Predict the anomaly temperature

After training the regressor, each regression model is fed with input of feature matrix.

```python
# concate the prediction from three algorithm to the dataset
dataset_tempe['LR prediction'] = lr_tempe.predict(X_tempe_train)
dataset_tempe['RFR prediction'] = regr_rf.predict(X_tempe_train)
dataset_tempe['SVR prediction'] = svr_rbf.predict(X_tempe_train)
print('Dataframe shape:', dataset_tempe.shape)
```

The prediction of each regressor is plotted and compared with the origin or 5 year mean global temperature anomaly plot.

```python
# display the results of comparison
result_columns = ['5 year mean', 'LR prediction', 'RFR prediction', 'SVR prediction']
dataset_tempe[result_columns].plot(figsize = (16,5), 
                                   alpha=0.75, 
                                   style = ['b-','r--','g:', 'c-.'],
                                   linewidth=3.0,
                                  )
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.title('Temperature anomaly prediction result comparison')
```

For further analysis and in order to quantify the regression performance, several regression metrics are computed and compared between three regression models. 

```python
# import regression metrics
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score

# list of metrics name
metricsList = ['Explained variance score','Mean absolute error','Mean squared error','Median absolute error',' R² score']

# list of prediction algorithm
predictNameList = ['LR prediction', 'RFR prediction', 'SVR prediction']

# list of metrics method
metricsFuncList  = [explained_variance_score, mean_absolute_error, mean_squared_error, median_absolute_error,r2_score]

# create a dataframe of metrics
df_metrics = pd.DataFrame(index=metricsList,columns=predictNameList)

# iterate over the metrics dataframe
for i in range(0,len(df_metrics)):
    for j in range(0,len(predictNameList)):
        df_metrics.iloc[i][j] = metricsFuncList[i](dataset_tempe[['5 year mean']],dataset_tempe[[predictNameList[j]]])

# show the results
df_metrics        
```

The table above shows model evaluation regarding the three implemented regression model, i.e  Linear Regression, Random Forest Regression and Support Vector Regression. Mean absolute error, mean squared error and median absolute error show the difference values between prediction and target, so the smaller the value, the prediction is better. As table above describes, Linear Regression prediction has the lowest value. 

Meanwhile, explained variance score measure the effect of data variance towards model and R<sup>2</sup> score measure how well future samples are likely to be predicted by the model. So the bigger the value of those two metrics, the prediction is better. As we can see, Linear Regression prediction still has the biggest score in both metrics. 

Thus, we will use **Linear Regression** to predict global temperature anomaly, as it has the best quality of prediction compared to other two methods.

```python
# copy the df values to the new extended input matrix
X_tempe_train_extend = X_tempe_train[X_tempe_train.columns.values]

# extend the input matrix based on number of years we want
start_pred = 2017
end_pred = 2020
duration = end_pred - start_pred + 1
years_prediction = list(range(start_pred, end_pred+1))

# convert the type of dataframe to numeric for interpolation purpose
for col in X_tempe_train_extend:
    X_tempe_train_extend[col] = pd.to_numeric(X_tempe_train_extend[col], errors='coerce')

# fill the extended matrix with nan
for year in years_prediction:
    X_tempe_train_extend.loc[str(year)] = np.nan
    
# check the shape of the dataframe shape
print('Dataframe shape:', X_tempe_train_extend.shape)
```

```python
# reset the index and the column name of the dataframe
X_tempe_train_extend_reset = X_tempe_train_extend.reset_index(drop=True)
X_tempe_train_extend_reset.columns = [i for i in range(X_tempe_train_extend.shape[1])]
```

These empty values of each indicatros from 2017 to 2020 is filled with interpolation values.

```python
# interpolate the indicator values based on the previous years data
X_tempe_train_extend_inter = X_tempe_train_extend_reset.interpolate(method='pchip',limit_direction='both',limit=duration,axis=0)

# check the number of NaN row
print('Number of NaN row:', X_tempe_train_extend_inter.isnull().any().sum())
```

```python
# giving the data frame original index and column name
dataset_tempe_extend = X_tempe_train_extend_inter[list(range(0,X_tempe_train_extend_inter.shape[1]))]
dataset_tempe_extend.columns = X_tempe_train_extend.columns
dataset_tempe_extend.index = X_tempe_train_extend.index
```

The extended feature matrix is fed to Linear Regression model to predict the global temperature in future.

```python
# predict the future global temperature by feeding the model with interpolated input matrix
dataset_tempe_extend['LR future prediction'] = lr_tempe.predict(X_tempe_train_extend_inter)
```

The predicted values are plotted and compared with original target values.

```python
# set figure size
plt.figure(figsize=(16,5))

# plot the prediction
plt.plot(dataset_tempe_extend.index.values, dataset_tempe_extend['LR future prediction'], ls = '-')

# plot the target vector
plt.plot(dataset_tempe.index, dataset_tempe['5 year mean'], ls = '--')

# set plot attributes
plt.xlabel("Year")
plt.ylabel('Temperature Anomaly (C)')
plt.title('Temperature anomaly future prediction comparison')
plt.legend()
```
