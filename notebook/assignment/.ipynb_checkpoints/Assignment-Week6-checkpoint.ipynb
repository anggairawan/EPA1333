## RMS Titanic 

One file is provided:
  * ```titanic.csv``` - this file contains titanic passenger information

The file ```titanic.csv``` provides passenger information on the ship RMS Titanic that sank on her maiden voyage on April 15, 1912. 

**1.** Read the passenger information of the Titanic. Show an overview of _unique_ last names of the passenger list (so ignore the first names and the "Mr", "Mrs.", etc) **and** determine the number of different last names.

````python
# Read the passenger information of the Titanic. 
df = pd.read_csv('titanic.csv', delimiter=',', skipinitialspace=True)
df # check the csv
````

````python
# Show an overview of unique last names of the passenger list (so ignore the first names and the "Mr", "Mrs.", etc) 
# take 'Name' column, split it by comma, take the first element, add unique so it will be the set of the last name
df_unique = pd.DataFrame(df['Name'].str.split(',').str[0].unique(),  columns=["Unique Last Name"])
df_unique #check the unique
````

````python
# and determine the number of different last names.
df_unique.count()
````

**2.** Assuming that passengers with the same last name form a family, determine how big the largest family aboard the Titanic was.

````python
# Assuming that passengers with the same last name form a family, 
# determine how big the largest family aboard the Titanic was.
df_family_number = df['Name'].str.split(',').str[0].value_counts()
max_number_of_person_in_family = df_family_number.max()

df_family_number [df_family_number == max_number_of_person_in_family]
````

**3.** Drop all the rows for which the Fare is 0 or NaN. Store the result.

````python
# create fare dataframe
df_fare =  df['Fare']

# replace the NaN with 0.0
df_fare.replace(np.nan, float(0), inplace = True)
#df_fare.isnull().sum() # check the NaN

# drop the row with fare == 0.0
df_drop_fare = df[df_fare != float(0)]
df_drop_fare
````

**4.** Show the average Fare price per Passenger Class (Pclass) in a bar chart.

````python
# check the available classes 
df_drop_fare['Pclass'].value_counts()

# obtain the average fare for every passenger class
df_pclass3 = df[df['Pclass'] == 3]
pc3_fare_mean = df_pclass3['Fare'].mean()
df_pclass2 = df[df['Pclass'] == 2]
pc2_fare_mean = df_pclass2['Fare'].mean()
df_pclass1 = df[df['Pclass'] == 1]
pc1_fare_mean = df_pclass1['Fare'].mean()

# create a dataframe for the average fare
pc_fare_mean = [pc1_fare_mean, pc2_fare_mean, pc3_fare_mean]
pc_fm_array = np.array(pc_fare_mean)
df_pcm = pd.DataFrame(pc_fm_array, columns = ['Pclass mean'], index = list('123'))
df_pcm.index.name = 'Passenger Class'

pcm_plot = df_pcm.plot(kind = 'bar', title = "Passenger Class Fare Average")
pcm_plot.set_ylabel("Fare")
````

**5.** Plot a histogram of the passenger's age, grouped by sex. So the end-result should show separate  histograms for males and females.

````python
# create a data frame of age for each sex
df_male = df[df['Sex'] == 'male']
df_male_age = df_male['Age']
df_female = df[df['Sex'] == 'female']
df_female_age = df_female['Age']

# calculate the histogram of both sex
df_ma_hist = df_male_age.value_counts()  
df_fa_hist = df_female_age.value_counts()

# sorting the histogram based on series's index (age of dataframe)
ma_hist_sorted = df_ma_hist.sort_index(ascending=True)
fa_hist_sorted = df_fa_hist.sort_index(ascending=True)

# rename the histogram column name
df_ma = ma_hist_sorted.to_frame().rename(columns={"Age": "Male Occurrence"})
df_fa = fa_hist_sorted.to_frame().rename(columns={"Age": "Female Occurrence"})

# Copy the index to the new column so we have age column
df_ma['Age'] = df_ma.index
df_fa['Age'] = df_fa.index

# reset the index [0,1,2...]
df_ma = df_ma.reset_index(drop=True)
df_fa = df_fa.reset_index(drop=True)

# merge two dataframe based on age, the outer will make it includes both non-match rows
df_hist = pd.merge(df_ma, df_fa, on="Age", how="outer")

# sorting the df based on the combined age
df_hist_sorted = df_hist.sort_values( by='Age', ascending=True )
# reordering the column name
df_hist_sorted = df_hist_sorted[['Age', 'Male Occurrence', 'Female Occurrence']]

# plot both histogram in one figure
df_hist_sorted.plot('Age',['Male Occurrence','Female Occurrence'], kind="bar", figsize=(16, 6), width = 1.2)
# axis label
plt.ylabel("Occurrence")
# Plot a title
plt.title("Histogram of titanic's' passenger age" )
````
